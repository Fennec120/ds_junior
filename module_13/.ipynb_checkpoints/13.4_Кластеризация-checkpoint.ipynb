```python
k = 3
x = x.values

np.random.seed(1)
idx = np.random.choice(x.shape[0], k, replace=False)
centroids = x[idx, :]

centroids
```

```python
tol = 1e-4
stop_criteria = False

while not stop_criteria:
    distances = cdist(x, centroids, 'euclidean')
    print(distances)
    points = distances.argmin(1)
    print(points)

    new_centroids = []

    for i in range(k):
        pts = x[points == i]
        new_centroids.append(pts.mean(axis=0))
  
    old_centroids = centroids.copy()
    centroids = np.vstack(new_centroids)
    stop_criteria = (np.abs(old_centroids - centroids) > tol).sum() == 0
```

```python
class MyKMeans(object):
    def __init__(self, k, tol=1e-4):
        self.k = k
        self.tol = tol
        self.centroids = None

    def fit(self, x):
        idx = np.random.choice(x.shape[0], self.k, replace=False)
        centroids = x[idx, :]
        stop_criteria = False

        while not stop_criteria:
            distances = cdist(x, centroids, 'euclidean')
            points = distances.argmin(1)

            new_centroids = []

            for i in range(self.k):
                pts = x[points == i]
                new_centroids.append(pts.mean(axis=0))

        old_centroids = centroids.copy()
        centroids = np.vstack(new_centroids)
        stop_criteria = (np.abs(old_centroids - centroids) > self.tol).sum() == 0

        self.centroids = centroids
        return self

    def predict(self, x):
        distances = cdist(x, self.centroids, 'euclidean')
        points = distances.argmin(1)
        return points
```

```python
x = df.iloc[:, :2].values
km = MyKMeans(k=3).fit(x)

point = km.predict(x)
```

```python
sns.scatterplot(df['Sum'], df['Days'], hue=point)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('data.csv')
df.head()
```

```python
km = KMeans(n_clusters=3)
x = df.iloc[:, :2]

km.fit(x)

points = km.predict(x)
plt.scatter(df['Sum'], df['Days'], c=df.Color)
```

```python
n_clusters = list(range(2, 10))

metrics = []

for k in n_clusters:
    km = KMeans(n_clusters=k).fit(x)
    metrics.append(km.inertia_)

plt.plot(n_clusters, metrics)
```
