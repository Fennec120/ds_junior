import sqlite3
from faker import Faker

fake = Faker()

# Establish connection to/create the database file (e.g., 'database.db')
conn = sqlite3.connect('data/database.db')
cursor = conn.cursor()

# Create the 'categories' table
cursor.execute("""
CREATE TABLE IF NOT EXISTS categories (
  id INTEGER PRIMARY KEY AUTOINCREMENT,
  name TEXT UNIQUE 
)
""")

# Create the 'products' table
cursor.execute("""
CREATE TABLE IF NOT EXISTS products (
  id INTEGER PRIMARY KEY AUTOINCREMENT,
  name TEXT,
  category_id INT,
  FOREIGN KEY (category_id) REFERENCES categories(id)
)
""")

# Generate random data for categories and insert them into the 'categories' table
for _ in range(15):
    category_name = fake.word().capitalize()
    cursor.execute("INSERT OR IGNORE INTO categories (name) VALUES (?)", (category_name,))
    
# Generate random data for products, assign multiple random categories per product,
# and insert them into the 'products' table
for _ in range(100):
    product_name = fake.name()
    
    # Get a list of randomly selected category IDs from existing entries in the 'categories' table.
    cursor.execute("SELECT id FROM categories ORDER BY RANDOM() LIMIT ?", (fake.random_int(min=1, max=5),))
    category_ids = [row[0] for row in cursor.fetchall()]
    
    if not category_ids:
        # If no categories are assigned, add this product to a separate group without any specific category.
        cursor.execute("INSERT INTO products (name) VALUES (?)", (product_name,))
        
    else:
        # Assign each randomly selected category ID to the product
        for category_id in category_ids:
            cursor.execute("INSERT INTO products (name, category_id) VALUES (?, ?)", (product_name, category_id))

conn.commit()
cursor.close()
conn.close()

```python

def calculate_iqr_boundaries(series):
    q25 = series.quantile(0.25)
    q75 = series.quantile(0.75)
    iqr = q75 - q25

    boundaries = (q25 - 1.5 * iqr, q75 + 1.5 * iqr)
    return boundaries



def bounds_f(x_train):
    bounds = []
    bounds_d = dict()
    for col in x_train.columns:
        if col == 'gender':
            continue
    #     iter_name = f'{col}'
        exec(f'{col}_boundaries = calculate_iqr_boundaries(x_train.{col})')
        exec(f'bounds.append(({col}_boundaries))')
    for i in range(len(x_train.columns)):
        if x_train.columns[i] == 'gender':
            continue
        bounds_d[x_train.columns[i]] = bounds[i-1]
    return bounds_d
```

```python
clusters = [2,3,4,5,6,7]
metrics = []
metrics_sil = []
ax = plt.subplots(1,2)[1]

for i in clusters:
    
    km = KMeans(n_clusters=i).fit(df)
    metrics.append(km.inertia_)
    metrics_sil.append(silhouette_score(df, km.labels_))

ax[0].plot(clusters, metrics)
ax[1].plot(clusters, metrics_sil)
```
