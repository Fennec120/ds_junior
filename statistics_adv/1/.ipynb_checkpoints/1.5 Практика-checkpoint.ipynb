# 1.4 Иллюстрация центральной предельной теоремы на Python

Прежде чем приступать к демонстрации ЦПТ, давайте вспомним, что это за теорема и какие выводы можно сделать.

Напомним формулировку $\textbf{центральной предельной теоремы}$.

Пусть $X_{1},\ldots ,X_{n},\ldots$  — бесконечная последовательность независимых одинаково распределённых случайных величин, имеющих конечное математическое ожидание $\mu$  и дисперсию $\sigma ^{2}$.

$\textbf{Выборочным средним}$ называется случайная величина $\quad {\displaystyle {\overline{X}}_{n}={\frac {1}{n}}\sum \limits _{i=1}^{n}X_{i}}. \quad$

Тогда $\quad \boxed{\displaystyle {\sqrt {n}}\cdot {\frac {{\overline{X}}_{n}-\mu }{\sigma }}\to N(0,1)} \quad$ **по распределению** при ${\displaystyle n\to \infty }$.

В формулировке ЦПТ используется фраза сходятся по распределению. Давайте дадим здесь определение для большей ясности:      
**Определение**: Случайные величины $X_n$ **сходятся по распределению** к случайной величине $X$, если их функции распределения $F_{X_n}$ сходятся поточечно к функции распределения предела $F_x$ во всех ее точках непрерывности.

В статистике, как и в математике, всегда пытаются однозначно дать какие-то определения, поэтому они могут казаться перегруженными и сложными. 
Хотя сама идея может быть очень простой:
- При большом количестве случайных величин их сумму хорошо приближает нормальное распределение;
- Именно этот факт активно используют в статистике для анализа сумм и выборочных средних.

Если вы и сейчас не очень поняли, что же делает эта теорема, то самое время перейти к примерам и на практике посмотреть, как она работает.

# 2. Пример для бернуллиевской случайной величины

```python
## задаем распределение с параметрами
rv_bernoulli = stats.bernoulli(p=0.5)
```

```python
X = [0, 1]

plt.plot(X, rv_bernoulli.pmf(X), ls="",marker="s",color='b')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Probability mass function (pmf)')
plt.show()
```

```python
# данная команда генерирует одну выборку длиной 10 из заданного распределения
rv_bernoulli.rvs(10)
```

```python
# данная команда генерирует уже 2 выборки длиной 10 из заданного распределения
y = rv_bernoulli.rvs((2, 10))
y
```

```python
np.mean(y, axis=0)
```

Нам понадобится функция np.mean. Данная функция позволяет посчитать среднее не только от массива, но и от более сложных вложенных структур. Обычно данные структуры являются матрицей. Поэтому среднее считается по осям - либо по набору чисел только по вертикали, либо только по горизонтали, либо вообще по всем элементам.

```python
# среднее вообще всех чисел 1,2,3,4
a = np.array([[1, 2], [3, 4]])
print(a.mean())
```

```python
# среднее 1,3 и 2,4
np.mean(a, axis=0)
```

```python
# среднее 1,2 и 3,4
np.mean(a, axis=1)
```

## 2.1 Задаем нормальную случайную величину

```python
# Сгенерируем станартное нормальное распределение
mu = 0
sigma = 1

# rv_norm - объект нормального распределения
rv_norm = stats.norm(mu, sigma)
```

```python
X = np.linspace(-4, 4, 100)
plt.plot(X, rv_norm.pdf(X))

plt.title('График функции плотности нормального распредления')
plt.xlabel('x')
plt.ylabel('p(x)')
plt.show()
```

```python
# математическое ожидание и дисперсия
rv_norm.mean(), rv_norm.var()
```

## 2.2 Функция для демонстрации

```python
def CLT_visualisation(X, n):
    '''
    X - объект, который является случайной величиной
    n - количество выборок, которое мы генерируем
    '''
    # математическое ожидание и среднеквадратическое отклонение
    mu, sigma = X.mean(), X.std() 
    
    # генерация n выборок большого размера
    sample = X.rvs((n, 10000)) 
    
    # считаем среднее арифметическое первых, вторых и тд элементов каждой из выборок
    X_n = sample.mean(axis=0) # среднее арифметическое
    
    # Формула из ЦПТ
    approximation = np.sqrt(n) * (X_n - mu) / sigma 
    # Визуализируем
    plt.hist(approximation, bins=50, density=True, label=f'n = {str(n)}')
    
    # График плотности нормального распределения
    Y = np.linspace(-4, 4, 100)
    plt.plot(Y, rv_norm.pdf(Y), linewidth=3, label='N(0, 1)')
    plt.legend()
```

## 2.3 Демонстрация ЦПТ для бернуллиевской случайной величины

```python
CLT_visualisation(rv_bernoulli, 1)
```

```python
plt.figure(1)
CLT_visualisation(rv_bernoulli, 3)

plt.figure(2)
CLT_visualisation(rv_bernoulli, 6)

plt.figure(3)
CLT_visualisation(rv_bernoulli, 10)

plt.figure(4)
CLT_visualisation(rv_bernoulli, 50)

plt.figure(5)
CLT_visualisation(rv_bernoulli, 100)

plt.figure(6)
CLT_visualisation(rv_bernoulli, 1000)
```

## 3. Демонтрация ЦПТ для показательного распределения

```python
# задаем показательное распределение
rv_exp = stats.expon(loc=0, scale=1)
```

```python
# мат ожидание и дисперсия
rv_exp.mean(), rv_exp.var()
```

```python
X = np.linspace(0,5,100)
plt.plot(X, rv_exp.cdf(X))

plt.title('График функции распределения')
plt.xlabel('x')
plt.ylabel('F(x)')
plt.show()
```

```python
X = np.linspace(0,5,100)
plt.plot(X, rv_exp.pdf(X))

plt.title('График функции плотности показательного распределения')
plt.xlabel('x')
plt.ylabel('p(x)')
plt.show()
```

```python
CLT_visualisation(rv_exp, 1)
```

```python
plt.figure(1)
CLT_visualisation(rv_exp, 3)

plt.figure(2)
CLT_visualisation(rv_exp, 6)

plt.figure(3)
CLT_visualisation(rv_exp, 10)

plt.figure(4)
CLT_visualisation(rv_exp, 50)

plt.figure(5)
CLT_visualisation(rv_exp, 100)
```

**Вывод:** Вот поэтому нормальное распределение очень часто используют, чтобы приближать различные явления, когда, например, рассчитывается какое-то среднее.
