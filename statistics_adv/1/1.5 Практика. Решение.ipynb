```python
# расширяем поле ноутбука для удобства
from IPython.core.display import display, HTML
display(HTML('<style>.container {width:95% !important;}</style>'))

# убираем предупреждения
import warnings
warnings.filterwarnings('ignore')
```

```python
# импортируем необходимые библиотеки
import numpy as np
import pandas as pd

import scipy
from scipy import stats

import matplotlib.pyplot as plt
import seaborn as sns

# позволяет рисовать графики внутри ноутбука
%matplotlib inline
plt.style.use('seaborn-whitegrid')
```

```python
# Сгенерируем станартное нормальное распределение
mu = 0
sigma = 1

# rv_norm - объект нормального распределения
rv_norm = stats.norm(mu, sigma)


def CLT_visualisation(X, n):
    '''
    X - объект, который является случайной величиной
    n - количество выборок, которое мы генерируем
    '''
    # математическое ожидание и среднеквадратическое отклонение
    mu, sigma = X.mean(), X.std() 
    
    # генерация n выборок большого размера
    sample = X.rvs((n, 10000)) 
    
    # считаем среднее арифметическое первых, вторых и тд элементов каждой из выборок
    X_n = sample.mean(axis=0) # среднее арифметическое
    
    # Формула из ЦПТ
    approximation = np.sqrt(n) * (X_n - mu) / sigma 
    # Визуализируем
    plt.hist(approximation, bins=50, density=True, label=f'n = {str(n)}')
    
    # График плотности нормального распределения
    Y = np.linspace(-4, 4, 100)
    plt.plot(Y, rv_norm.pdf(Y), linewidth=3, label='N(0, 1)')
    plt.legend()
```

```python
## необходимо задать равномерное распределение

a, b = 0, 1
rv_uniform = stats.uniform(loc=a, scale=b-a)
```

```python
## вывести график функции плотности распределения

X = np.linspace(-0.5, 1.5, 100)
plt.plot(X, rv_uniform.pdf(X))

plt.title('График функции плотности показательного распределения')
plt.xlabel('x')
plt.ylabel('p(x)')
plt.show()
```

```python
## вывести математическое ожидание и дисперсию

rv_uniform.mean(), rv_uniform.var()
```

```python
## применить функцю CLT_visualisation и проанализировать результат

plt.figure(1)
CLT_visualisation(rv_uniform, 1)

plt.figure(2)
CLT_visualisation(rv_uniform, 3)

plt.figure(3)
CLT_visualisation(rv_uniform, 6)

plt.figure(4)
CLT_visualisation(rv_uniform, 10)

plt.figure(5)
CLT_visualisation(rv_uniform, 50)

plt.figure(6)
CLT_visualisation(rv_uniform, 100)
```

**Вывод**: сначала при маленьком количестве случайных величин из равномерного распределения мы получаем фактически само распределение. При увеличении количества случайных вличин мы все больше и больше получаем распределения, похожее на нормальное. По ЦПТ это и есть нормальное распределение.
